### Get unique store location list

```python
store_loc_df = sales_df[['Store Latitude','Store Longitude']].drop_duplicates()
```

```python
store_list_u = store_loc_df.values.tolist()
store_list_u
```

### Convert coordinates for map

```python
from pyproj import Proj, transform
```

#### change stores' latitude/longitude to web_cor scale

```python
store_cor = []
for (s_lat,s_lon) in store_list_u:
        store_cor.append(transform(Proj(init='epsg:4326'), Proj(init='epsg:3857'), s_lon, s_lat))
store_cor
```

```python
store_cor
```

```python
store_xcor = [x[0] for x in store_cor]
store_ycor = [x[1] for x in store_cor]
```

```python
len(store_cor)
```

#### Define functions to get web_cor for store

```python
def s_x_cord(num):
    s_df = sales_df.loc[(sales_df['Store Latitude'] == store_list_u[num][0]) & 
                     (sales_df['Store Longitude'] == store_list_u[num][1])]
    c_df = s_df[['Customer Latitude', 'Customer Longitude']].drop_duplicates()
    c_list_u = c_df.values.tolist()
    c_cor = []
    for (c_lat,c_lon) in c_list_u:
        c_cor.append(transform(Proj(init='epsg:4326'), Proj(init='epsg:3857'), c_lon, c_lat))
    xlist = [a for (a,b) in c_cor]
    return xlist
def s_y_cord(num):
    s_df = sales_df.loc[(sales_df['Store Latitude'] == store_list_u[num][0]) & 
                     (sales_df['Store Longitude'] == store_list_u[num][1])]
    c_df = s_df[['Customer Latitude', 'Customer Longitude']].drop_duplicates()
    c_list_u = c_df.values.tolist()
    c_cor = []
    for (c_lat,c_lon) in c_list_u:
        c_cor.append(transform(Proj(init='epsg:4326'), Proj(init='epsg:3857'), c_lon, c_lat))
    ylist = [b for (a,b) in c_cor]
    return ylist
```

```python
store_idx = 0
x_cord_list = []
while store_idx < 15:
    x_cord = s_x_cord(store_idx)
    x_cord_list.append(x_cord)
    store_idx=store_idx+1
```

```python
store_idx = 0
y_cord_list = []
while store_idx < 15:
    y_cord = s_y_cord(store_idx)
    y_cord_list.append(y_cord)
    store_idx=store_idx+1
```

```python
len(x_cord_list)
```

```python
test_x = [ b for a in x_cord_list for b in a]
```

```python
test_y = [ b for a in y_cord_list for b in a]
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.tile_providers import CARTODBPOSITRON_RETINA

output_notebook()

size_set =  [8]*2499 +[14]*15 
color_set = ['#33ccff']*2499  + ['cc0000']*15
alpha_set = [0.2]*2499 + [0.8]*15

ttl_x2 = test_x + store_xcor
ttl_y2 = test_y + store_ycor

x= ttl_x2
y= ttl_y2

links = {
    2499: [i for i in range(0,375)],
    2500: [i for i in range(375,375+106)],
    2501: [i for i in range(375+106,375+106+258)],
    2502: [i for i in range(375+106+258,375+106+258+366)],
    2503: [i for i in range(375+106+258+366,375+106+258+366+140)],
    2504: [i for i in range(375+106+258+366+140,375+106+258+366+140+117)],
    2505: [i for i in range(375+106+258+366+140+117,375+106+258+366+140+117+66)],
    2506: [i for i in range(375+106+258+366+140+117+66,375+106+258+366+140+117+66+264)],
    2507: [i for i in range(375+106+258+366+140+117+66+264,375+106+258+366+140+117+66+264+56)],
    2508: [i for i in range(375+106+258+366+140+117+66+264+56,375+106+258+366+140+117+66+264+56+520)],
    2509: [i for i in range(375+106+258+366+140+117+66+264+56+520,375+106+258+366+140+117+66+264+56+520+32)],
    2510: [i for i in range(375+106+258+366+140+117+66+264+56+520+32,375+106+258+366+140+117+66+264+56+520+32+25)],
    2511: [i for i in range(375+106+258+366+140+117+66+264+56+520+32+25,375+106+258+366+140+117+66+264+56+520+32+25+98)],
    2512: [i for i in range(375+106+258+366+140+117+66+264+56+520+32+25+98,375+106+258+366+140+117+66+264+56+520+32+25+98+37)],
    2513: [i for i in range(375+106+258+366+140+117+66+264+56+520+32+25+98+37,375+106+258+366+140+117+66+264+56+520+32+25+98+37+39)]
}


TOOLS="pan,wheel_zoom,box_zoom,reset"
# range bounds supplied in web mercator coordinates
p = figure(x_range=(-15000,2000), y_range=(6675000,6750000),
           x_axis_type="mercator", y_axis_type="mercator", plot_width=980, plot_height=700,
           tools=TOOLS, toolbar_location='right')

p.axis.visible = False

p.add_tile(CARTODBPOSITRON_RETINA)

# Title
p.title.text = "Store and Customer Connection Map"
p.title.align = "center"
p.title.text_color = "black"
p.title.text_font_size = "20px"

source = ColumnDataSource({'x0': [], 'y0': [], 'x1': [], 'y1': []})
cr = p.circle(x, y, color=color_set, size=size_set, alpha=alpha_set, hover_color = '#990000', hover_alpha=1.0)
sr = p.segment(x0='x0', y0='y0', x1='x1', y1='y1', color='#990000', alpha=0.3, line_width=1.2, source=source, )


# Add a hover tool, that sets the link data for a hovered circle
code = """
var links = %s;
var data = {'x0': [], 'y0': [], 'x1': [], 'y1': []};
var cdata = circle.data;
var indices = cb_data.index['1d'].indices;
for (var i = 0; i < indices.length; i++) {
    var ind0 = indices[i]
    for (var j = 0; j < links[ind0].length; j++) {
        var ind1 = links[ind0][j];
        data['x0'].push(cdata.x[ind0]);
        data['y0'].push(cdata.y[ind0]);
        data['x1'].push(cdata.x[ind1]);
        data['y1'].push(cdata.y[ind1]);
    }
}
segment.data = data;
""" % links

callback = CustomJS(args={'circle': cr.data_source, 'segment': sr.data_source}, code=code)
p.add_tools(HoverTool(tooltips=None, callback=callback, renderers=[cr]))

show(p)


## Some hints while using above map

#  Please do not turn off the "hover" button in the tool bar
#  You can use "Box Zoom" and "Wheel Zoom" in the tool bar to zoom up the map if you are interested in some specific area. Also you can click the "Reset" to return the default view of the map.
#  Have fun!
```
